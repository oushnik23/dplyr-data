**Remove duplicate rows based on Index column in my_data and assign data as data1.**

```python
data1=data.drop_duplicates(['Index'])
```

```python
data1
```

```python
data1.shape
```

**Drop variables name starts with Y in my_data and assign as data2**

```python
data2=data.loc[:,~data.columns.str.startswith('Y')]
```

```python
data2
```

**select variables in my_data ,contain I in their name and assign as data3.**

```python
data3=data.loc[:,data.columns.str.contains('I')]
data3
```

**Filter rows in my_data where Index=A or C and assign as data4.**

```python
data4=data[data['Index'].isin(['A','C'])]
data4
```

**Filter rows in my_data where Index !=A or C and assign as data5.**

```python
data5=data[data['Index'].isin(['A','C'])==False]
data5
```

**Get the mean and median of Y2011 column based on Index column in my_data.**

```python
data.groupby('Index')['Y2011'].mean().sort_values(ascending=False)
```

```python
data.groupby('Index')['Y2011'].median().sort_values(ascending=False)
```

**select 10 % rows randomly from my_data and assign as data6**

```python
data6=data.sample(frac=.10)
data6
```

**Export the data6 as my_new_data(in csv) in your working directory.**

```python
data6.to_csv('my_new_data.csv',index=False)
```
